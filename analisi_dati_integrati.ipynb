```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulazione del dataset integrato
data_integrato = {
    "Nome": ["Hotel Inclusivo Roma", "B&B Accessibile Firenze", "Agriturismo Senza Barriere Napoli",
             "Hotel Accessibile Milano", "B&B Accogliente Torino"],
    "Posizione": ["Lazio, Roma", "Toscana, Firenze", "Campania, Napoli", "Lombardia, Milano", "Piemonte, Torino"],
    "Regione": ["Lazio", "Toscana", "Campania", "Lombardia", "Piemonte"],
    "Turisti_accessibili": [500000, 540000, 256000, 900000, 270000],
    "Turisti_totali": [5000000, 4500000, 3200000, 6000000, 3000000]
}

df_integrato = pd.DataFrame(data_integrato)

# Distribuzione dei turisti accessibili per regione
plt.figure(figsize=(10, 6))
df_integrato.groupby("Regione")["Turisti_accessibili"].sum().plot(
    kind="bar", color="skyblue", title="Distribuzione dei Turisti Accessibili per Regione"
)
plt.ylabel("Numero di Turisti Accessibili")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Rapporto tra strutture disponibili e turisti accessibili
df_integrato["Rapporto_strutture_turisti"] = (
    df_integrato["Turisti_accessibili"] / df_integrato.groupby("Regione")["Nome"].transform("count")
)

plt.figure(figsize=(10, 6))
df_integrato.groupby("Regione")["Rapporto_strutture_turisti"].mean().plot(
    kind="bar", color="salmon", title="Rapporto Strutture Disponibili vs Turisti Accessibili"
)
plt.ylabel("Rapporto Strutture / Turisti Accessibili")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
